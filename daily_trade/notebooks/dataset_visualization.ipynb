## 1. „É©„Ç§„Éñ„É©„É™„ÅÆ„Ç§„É≥„Éù„Éº„Éà


```python
# ÂøÖË¶Å„Å™„É©„Ç§„Éñ„É©„É™„ÅÆ„Ç§„É≥„Éù„Éº„Éà
import pandas as pd
import plotly.offline as pyo
import warnings

# Ë≠¶Âëä„ÅÆÈùûË°®Á§∫
warnings.filterwarnings("ignore")

# Plotly„ÅÆË®≠ÂÆö
pyo.init_notebook_mode(connected=True)

# „Éë„É≥„ÉÄ„Çπ„ÅÆË°®Á§∫Ë®≠ÂÆö
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", 50)
```

## 2. „Éá„Éº„Çø„Çª„ÉÉ„Éà„ÅÆË™≠„ÅøËæº„Åø„Å®Âü∫Êú¨ÊÉÖÂ†±Á¢∫Ë™ç


```python
# „Éá„Éº„Çø„Çª„ÉÉ„Éà„ÅÆ„Éë„Çπ„ÇíÊåáÂÆöÔºàÂøÖË¶Å„Å´Âøú„Åò„Å¶Â§âÊõ¥„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºâ
data_path = "../data/ohlcv/latest_dataset.parquet"

# „Éá„Éº„Çø„Çª„ÉÉ„Éà„ÅÆË™≠„ÅøËæº„Åø
df = None
used_path = None


df = pd.read_parquet(data_path)
used_path = data_path


# Âü∫Êú¨ÊÉÖÂ†±„ÅÆË°®Á§∫
print("\nüìä „Éá„Éº„Çø„Çª„ÉÉ„ÉàÂü∫Êú¨ÊÉÖÂ†±")
print(f"üìÅ ‰ΩøÁî®„Éï„Ç°„Ç§„É´: {used_path}")
print(f"üìè „Éá„Éº„ÇøÂΩ¢Áä∂: {df.shape}")
print(f"üíæ „É°„É¢„É™‰ΩøÁî®Èáè: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print(f"üìÖ ÊúüÈñì: {df['timestamp'].min()} „Äú {df['timestamp'].max()}")

# „Éá„Éº„ÇøÂûã„ÅÆÁ¢∫Ë™ç
print("\nüìã „Éá„Éº„ÇøÂûã:")
print(df.dtypes.value_counts())

# ÂÖàÈ†≠„Éá„Éº„Çø„ÅÆÁ¢∫Ë™ç
print("\nüëÄ „Éá„Éº„Çø„Çµ„É≥„Éó„É´:")
df.head()
```

## 3. ÈäòÊüÑ‰∏ÄË¶ß„Å®ÊúüÈñì„ÅÆÁ¢∫Ë™ç


```python
# ÈäòÊüÑ‰∏ÄË¶ß„ÅÆÂèñÂæó
symbols = df["symbol"].unique()
n_symbols = len(symbols)

print(f"üè¢ ÈäòÊüÑÊï∞: {n_symbols}")
print(f"üìù ÈäòÊüÑ‰∏ÄË¶ß: {', '.join(symbols)}")

# ÂêÑÈäòÊüÑ„ÅÆÁµ±Ë®àÊÉÖÂ†±
symbol_stats = (
    df.groupby("symbol").agg({"timestamp": ["min", "max", "count"], "close": ["mean", "std", "min", "max"]}).round(2)
)

symbol_stats.columns = ["ÈñãÂßãÊó•", "ÁµÇ‰∫ÜÊó•", "„Éá„Éº„ÇøÊï∞", "Âπ≥ÂùáÁµÇÂÄ§", "ÁµÇÂÄ§Ê®ôÊ∫ñÂÅèÂ∑Æ", "ÊúÄ‰ΩéÁµÇÂÄ§", "ÊúÄÈ´òÁµÇÂÄ§"]

print(f"\nüìä ÈäòÊüÑÂà•Áµ±Ë®à:")
print(symbol_stats)

# „Éá„Éº„ÇøÊúüÈñì„ÅÆÂèØË¶ñÂåñ
fig = px.timeline(
    df.groupby("symbol")["timestamp"].agg(["min", "max"]).reset_index(),
    x_start="min",
    x_end="max",
    y="symbol",
    title="üìÖ ÈäòÊüÑÂà•„Éá„Éº„ÇøÊúüÈñì",
    color="symbol",
    height=max(400, n_symbols * 30),
)

fig.update_layout(xaxis_title="ÊúüÈñì", yaxis_title="ÈäòÊüÑ", showlegend=False)

fig.show()

# ÂêÑÈäòÊüÑ„ÅÆ„Éá„Éº„ÇøÊï∞„ÅÆÂèØË¶ñÂåñ
data_counts = df["symbol"].value_counts().sort_values(ascending=True)

fig = px.bar(
    x=data_counts.values,
    y=data_counts.index,
    orientation="h",
    title="üìä ÈäòÊüÑÂà•„Éá„Éº„ÇøÊï∞",
    labels={"x": "„Éá„Éº„ÇøÊï∞", "y": "ÈäòÊüÑ"},
    color=data_counts.values,
    color_continuous_scale="viridis",
)

fig.update_layout(height=max(400, n_symbols * 25))
fig.show()
```

## 4. Âçò‰∏ÄÈäòÊüÑ„ÅÆÈÅ∏Êäû„Å®„Éï„Ç£„É´„Çø„É™„É≥„Ç∞


```python
# ÂàÜÊûêÂØæË±°ÈäòÊüÑ„ÅÆÈÅ∏ÊäûÔºàÂøÖË¶Å„Å´Âøú„Åò„Å¶Â§âÊõ¥„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºâ
# ‰∏ÄËà¨ÁöÑ„Å™ÈäòÊüÑ„ÇíÂÑ™ÂÖàÁöÑ„Å´ÈÅ∏Êäû
preferred_symbols = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "TSLA", "META"]
target_symbol = None

# Âà©Áî®ÂèØËÉΩ„Å™ÈäòÊüÑ„Åã„ÇâÂÑ™ÂÖàÈ†Ü‰Ωç„ÅßÈÅ∏Êäû
for symbol in preferred_symbols:
    if symbol in symbols:
        target_symbol = symbol
        break

# ÂÑ™ÂÖàÈäòÊüÑ„Åå„Å™„ÅÑÂ†¥Âêà„ÅØÊúÄÂàù„ÅÆÈäòÊüÑ„ÇíÈÅ∏Êäû
if target_symbol is None:
    target_symbol = symbols[0]

print(f"üéØ ÂàÜÊûêÂØæË±°ÈäòÊüÑ: {target_symbol}")

# Âçò‰∏ÄÈäòÊüÑ„Éá„Éº„Çø„ÅÆÊäΩÂá∫
symbol_df = df[df["symbol"] == target_symbol].copy()
symbol_df = symbol_df.sort_values("timestamp").reset_index(drop=True)

print(f"üìè {target_symbol} „Éá„Éº„ÇøÊï∞: {len(symbol_df)}")
print(f"üìÖ ÊúüÈñì: {symbol_df['timestamp'].min()} „Äú {symbol_df['timestamp'].max()}")
print(f"üóìÔ∏è  Âñ∂Ê•≠Êó•Êï∞: {len(symbol_df)}Êó•")

# ÊôÇÁ≥ªÂàó„ÅÆÈÄ£Á∂öÊÄßÁ¢∫Ë™ç
symbol_df["date"] = pd.to_datetime(symbol_df["timestamp"]).dt.date
date_gaps = pd.to_datetime(symbol_df["timestamp"]).diff().dt.days

print(f"\nüìä ÊôÇÁ≥ªÂàóÈÄ£Á∂öÊÄß:")
print(f"- Âπ≥ÂùáÊó•Êï∞ÈñìÈöî: {date_gaps.mean():.2f}Êó•")
print(f"- ÊúÄÂ§ßÊó•Êï∞ÈñìÈöî: {date_gaps.max():.0f}Êó•")
print(f"- 7Êó•‰ª•‰∏ä„ÅÆÁ©∫ÁôΩÊúüÈñì: {(date_gaps > 7).sum()}Âõû")

# Âü∫Êú¨Áµ±Ë®à„ÅÆË°®Á§∫
print(f"\nüìà {target_symbol} ‰æ°Ê†ºÁµ±Ë®à:")
price_stats = symbol_df[["open", "high", "low", "close", "volume"]].describe()
print(price_stats.round(2))
```

## 9. „Éá„Éº„ÇøÂìÅË≥™„ÉÅ„Çß„ÉÉ„ÇØ


```python
# ÊôÇÁ≥ªÂàóÈÄ£Á∂öÊÄß„ÅÆË©≥Á¥∞„ÉÅ„Çß„ÉÉ„ÇØ
print("‚è∞ ÊôÇÁ≥ªÂàóÈÄ£Á∂öÊÄß„ÉÅ„Çß„ÉÉ„ÇØ")
print("=" * 50)

# ÂêÑÈäòÊüÑ„ÅÆÊó•‰ªòÈñìÈöî„ÉÅ„Çß„ÉÉ„ÇØ
continuity_issues = []

for symbol in symbols:
    symbol_data = df[df["symbol"] == symbol].sort_values("timestamp")

    if len(symbol_data) > 1:
        date_diffs = pd.to_datetime(symbol_data["timestamp"]).diff().dt.days

        # 7Êó•‰ª•‰∏ä„ÅÆÈñìÈöî„ÇíÁï∞Â∏∏„Å®„Åø„Å™„Åô
        large_gaps = date_diffs[date_diffs > 7]

        if len(large_gaps) > 0:
            continuity_issues.append(
                {
                    "symbol": symbol,
                    "gaps_count": len(large_gaps),
                    "max_gap": large_gaps.max(),
                    "total_records": len(symbol_data),
                }
            )

if continuity_issues:
    continuity_df = pd.DataFrame(continuity_issues)
    print("‚ö†Ô∏è  ÊôÇÁ≥ªÂàóÈÄ£Á∂öÊÄß„ÅÆÂïèÈ°å„ÅåË¶ã„Å§„Åã„Çä„Åæ„Åó„Åü:")
    print(continuity_df)

    # ÈÄ£Á∂öÊÄßÂïèÈ°å„ÅÆÂèØË¶ñÂåñ
    fig_continuity = px.bar(
        continuity_df,
        x="symbol",
        y="gaps_count",
        title="üìä ÈäòÊüÑÂà•ÊôÇÁ≥ªÂàóÈñìÈöîÂïèÈ°åÊï∞ (7Êó•‰ª•‰∏ä„ÅÆÈñìÈöî)",
        labels={"gaps_count": "ÂïèÈ°åÊï∞", "symbol": "ÈäòÊüÑ"},
    )
    fig_continuity.show()
else:
    print("‚úÖ ÊôÇÁ≥ªÂàóÈÄ£Á∂öÊÄß„Å´ÂïèÈ°å„ÅØË¶ã„Å§„Åã„Çä„Åæ„Åõ„Çì„Åß„Åó„Åü")

# Áï∞Â∏∏ÂÄ§„ÅÆÊ§úÂá∫ÔºàÂçò‰∏ÄÈäòÊüÑÔºâ
print(f"\nüìä {target_symbol} Áï∞Â∏∏ÂÄ§Ê§úÂá∫")
print("=" * 50)

# ‰æ°Ê†º„Éá„Éº„Çø„ÅÆÁï∞Â∏∏ÂÄ§Ê§úÂá∫
price_columns = ["open", "high", "low", "close"]
available_price_cols = [col for col in price_columns if col in symbol_df.columns]

if available_price_cols:
    # Z-score„Å´„Çà„ÇãÁï∞Â∏∏ÂÄ§Ê§úÂá∫
    from scipy import stats

    anomalies_summary = []

    for col in available_price_cols:
        z_scores = np.abs(stats.zscore(symbol_df[col].dropna()))
        anomalies = z_scores > 3  # Z-score > 3„ÇíÁï∞Â∏∏ÂÄ§„Å®„Åô„Çã

        anomalies_summary.append(
            {
                "column": col,
                "anomalies_count": anomalies.sum(),
                "anomalies_percent": (anomalies.sum() / len(z_scores)) * 100,
            }
        )

    anomalies_df = pd.DataFrame(anomalies_summary)
    print("‰æ°Ê†º„Éá„Éº„ÇøÁï∞Â∏∏ÂÄ§ÂàÜÊûê (Z-score > 3):")
    print(anomalies_df.round(2))

    # „É™„Çø„Éº„É≥Áéá„ÅÆÁï∞Â∏∏ÂÄ§Á¢∫Ë™ç
    if "ret_1d" in symbol_df.columns:
        returns = symbol_df["ret_1d"].dropna()
        return_z_scores = np.abs(stats.zscore(returns))
        return_anomalies = return_z_scores > 3

        print(f"\n„É™„Çø„Éº„É≥ÁéáÁï∞Â∏∏ÂÄ§:")
        print(f"- Áï∞Â∏∏ÂÄ§Êï∞: {return_anomalies.sum()}")
        print(f"- Áï∞Â∏∏ÂÄ§Áéá: {(return_anomalies.sum() / len(returns)) * 100:.2f}%")

        if return_anomalies.sum() > 0:
            anomaly_dates = symbol_df.loc[returns.index[return_anomalies], "timestamp"]
            anomaly_returns = returns[return_anomalies]

            print("\nÁï∞Â∏∏„É™„Çø„Éº„É≥Ë©≥Á¥∞:")
            for date, ret in zip(anomaly_dates, anomaly_returns):
                print(f"- {date.strftime('%Y-%m-%d')}: {ret * 100:.2f}%")

# „Éá„Éº„Çø„ÅÆÁµ±Ë®à„Çµ„Éû„É™„Éº
print(f"\nüìà {target_symbol} Áµ±Ë®à„Çµ„Éû„É™„Éº")
print("=" * 50)

# Êï∞ÂÄ§Âàó„ÅÆÁµ±Ë®à
numeric_cols = symbol_df.select_dtypes(include=[np.number]).columns
stats_summary = symbol_df[numeric_cols].describe()

print("‰∏ªË¶ÅÁµ±Ë®àÈáè:")
key_cols = (
    ["close", "volume", "ret_1d", "rsi_14"]
    if all(col in numeric_cols for col in ["close", "volume", "ret_1d", "rsi_14"])
    else numeric_cols[:4]
)
available_key_cols = [col for col in key_cols if col in numeric_cols]

if available_key_cols:
    print(stats_summary[available_key_cols].round(4))

# „Éá„Éº„ÇøÂìÅË≥™„Çπ„Ç≥„Ç¢„ÅÆË®àÁÆó
print(f"\nüèÜ „Éá„Éº„ÇøÂìÅË≥™Ë©ï‰æ°")
print("=" * 50)

quality_score = 100
deductions = []

# Ê¨†ÊêçÂÄ§„Å´„Çà„ÇãÊ∏õÁÇπ
missing_rate = (symbol_df.isnull().sum().sum() / (len(symbol_df) * len(symbol_df.columns))) * 100
if missing_rate > 0:
    deduction = min(missing_rate * 2, 30)  # ÊúÄÂ§ß30ÁÇπÊ∏õÁÇπ
    quality_score -= deduction
    deductions.append(f"Ê¨†ÊêçÂÄ§ ({missing_rate:.2f}%): -{deduction:.1f}ÁÇπ")

# Áï∞Â∏∏ÂÄ§„Å´„Çà„ÇãÊ∏õÁÇπ
if "anomalies_df" in locals() and not anomalies_df.empty:
    avg_anomaly_rate = anomalies_df["anomalies_percent"].mean()
    if avg_anomaly_rate > 0:
        deduction = min(avg_anomaly_rate * 5, 20)  # ÊúÄÂ§ß20ÁÇπÊ∏õÁÇπ
        quality_score -= deduction
        deductions.append(f"Áï∞Â∏∏ÂÄ§ ({avg_anomaly_rate:.2f}%): -{deduction:.1f}ÁÇπ")

# ÈÄ£Á∂öÊÄß„Å´„Çà„ÇãÊ∏õÁÇπ
if continuity_issues:
    symbol_issues = [issue for issue in continuity_issues if issue["symbol"] == target_symbol]
    if symbol_issues:
        gap_rate = (symbol_issues[0]["gaps_count"] / symbol_issues[0]["total_records"]) * 100
        deduction = min(gap_rate * 10, 25)  # ÊúÄÂ§ß25ÁÇπÊ∏õÁÇπ
        quality_score -= deduction
        deductions.append(f"ÊôÇÁ≥ªÂàóÈñìÈöî ({gap_rate:.2f}%): -{deduction:.1f}ÁÇπ")

print(f"„Éá„Éº„ÇøÂìÅË≥™„Çπ„Ç≥„Ç¢: {quality_score:.1f}/100")
if deductions:
    print("Ê∏õÁÇπÂÜÖË®≥:")
    for deduction in deductions:
        print(f"  - {deduction}")
else:
    print("‚úÖ Ê∏õÁÇπË¶ÅÁ¥†„Å™„Åó - È´òÂìÅË≥™„Éá„Éº„Çø")

# ÂìÅË≥™Ë©ï‰æ°„ÅÆÂèØË¶ñÂåñ
quality_categories = ["ÂÆåÂÖ®ÊÄß", "Ê≠£Á¢∫ÊÄß", "ÈÄ£Á∂öÊÄß", "‰∏ÄË≤´ÊÄß"]
quality_scores = [
    100 - (missing_rate * 2),
    100 - (avg_anomaly_rate * 5 if "avg_anomaly_rate" in locals() else 0),
    100 - (gap_rate * 10 if continuity_issues else 100),
    quality_score,  # Á∑èÂêà„Çπ„Ç≥„Ç¢
]

fig_quality = go.Figure(
    data=go.Scatterpolar(r=quality_scores, theta=quality_categories, fill="toself", name=target_symbol)
)

fig_quality.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 100])),
    title=f"üèÜ {target_symbol} „Éá„Éº„ÇøÂìÅË≥™„É¨„Éº„ÉÄ„Éº„ÉÅ„É£„Éº„Éà",
    height=500,
)

fig_quality.show()
```
